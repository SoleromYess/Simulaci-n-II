#**Pruebas para determinar estacionariedad**

La prueba de ***Dickey-Fuller*** es una herramienta estadística que se utiliza para determinar ***si una serie de tiempo es estacionaria o no***. En otras palabras, nos ayuda a saber ***si la serie tiene una media y una varianza que se mantienen constantes en el tiempo***.

La prueba de Dickey-Fuller compara la serie de tiempo con una versión de esa misma serie que ha sido "diferenciada". La diferenciación implica calcular la diferencia entre un valor y el valor anterior en la serie.

Si la serie de tiempo original tiene raíces unitarias, lo que significa que los valores pasados afectan significativamente a los valores futuros, entonces la serie no es estacionaria..

Una raíz unitaria en el análisis de series de tiempo se refiere a una raíz de una ecuación característica que tiene un valor absoluto igual a 1. En el contexto del test de Dickey-Fuller, una serie de tiempo se considera tener raíces unitarias si las diferencias entre los valores adyacentes de la serie no se desvanecen a medida que aumenta el número de períodos. Esto indica que los efectos de los valores pasados en la serie de tiempo no se desvanecen a lo largo del tiempo y, por lo tanto, la serie no es estacionaria.

La prueba de Dickey-Fuller Aumentada (ADF) es una extensión de la prueba de Dickey-Fuller que incluye términos adicionales para tener en cuenta posibles tendencias y autocorrelaciones en la serie temporal.
es especialmente útil cuando se trabaja con series de tiempo que pueden tener tendencias lineales o no lineales.

Al implementar las pruebas de Dickey-Fuller y Dickey-Fuller permitirá tomar decisiones informadas sobre qué enfoques de modelado y pronóstico utilizar, así como realizar transformaciones en los datos si es necesario para lograr la estacionariedad.

#**Dickey-Fuller (DF)**

La Prueba Dickey-Fuller (DF) es una prueba estadística utilizada para evaluar la presencia de **raíces unitarias** y determinar **si una serie de tiempo es estacionaria o no**. Se utiliza para determinar si es necesario aplicar técnicas de diferenciación para hacerla estacionaria.


Fue desarrollada por los economistas David Dickey y Wayne Fuller en 1979.

La Prueba Dickey-Fuller básica (DF) se basa en un modelo autorregresivo de orden 1 *(AR(1))*.

El modelo autorregresivo de orden 1 se utiliza para analizar y predecir series de tiempo en las que la dependencia a largo plazo se puede explicar principalmente por el valor inmediatamente anterior. Sin embargo, si la serie de tiempo muestra una dependencia más compleja y de largo alcance, se pueden considerar modelos autorregresivos de orden superior.

El modelo DF se define como:

$$y(t) = \rho * y(t-1) + ϵ(t)$$

**Donde:**

*   $y(t)$ es el valor de la serie de tiempo en el tiempo $t$.
*   $ρ$ es el coeficiente de autorregresión.
*   $ε(t)$ es un término de error con media cero.


**Hipótesis:**

**$H_0:$** $ρ = 0.05$, la serie de tiempo es no estacionaria (Sus propiedades estadísticas cambian con el tiempo,es decir, la media, la varianza y la covarianza de la serie de tiempo se mantienen constantes a lo largo del tiempo) y tiene una tendencia determinista. La serie de tiempo tiene raíces unitarias.


**$H_A:$** $ρ < 0.05$ por lo que, la serie de tiempo no tiene raíces unitarias, entonces la serie de tiempo es estacionaria.

##**Dicker-Fuller: Procedimiento**

1. **Definir las hipótesis:**

**$H_0:$** $ρ = 0.05$, la serie de tiempo tiene raíz unitaria, es decir, no es estacionaria.

**$H_A:$** $ρ < 0.05$, la serie de tiempo no tiene raíz unitaria, es estacionaria.

2. **Calcular la serie de diferencias:**

La prueba Dickey-Fuller se realiza típicamente en la serie de primeras diferencias de la serie temporal original. La serie de diferencias se obtiene restando cada observación de la siguiente observación en la serie.

3. **Estimar el modelo de regresión:**

El primer paso es estimar un modelo de regresión para evaluar si la serie de diferencias tiene raíces unitarias. El modelo de regresión tiene la siguiente forma: $$Δy(t) = ρy(t-1) + α + β(t) + ε(t)$$

donde:

$Δy(t)$ representa la serie de diferencias

$ρ$  es el coeficiente que se prueba para determinar si es igual a cero (raíz unitaria).

$t$ es el tiempo

$α$ es una constante.

$β$ es un coeficiente que captura la tendencia de la serie.

$ε(t)$ es el término de error.

4. **Estimar el modelo de regresión:**

Se utilizan técnicas de estimación de mínimos cuadrados ordinarios (MCO) para estimar los coeficientes $ρ$, $α$ y $β$ del modelo de regresión.

5. **Calcular el estadístico de prueba:**

Se calcula el estadístico de prueba de Dickey-Fuller, que es una medida de cuánto se desvía $ρ$ de cero.
El estadístico de prueba se calcula de la siguiente manera:
Dickey-Fuller $$t = \frac {(ρ - 1)}{ std _{error}(ρ)}$$
$std_{error}(ρ)$ es el error estándar de la estimación del coeficiente $ρ$.

6. **Comparar el estadístico de prueba con los valores críticos:**

Se compara el valor del estadístico de prueba con los valores críticos de la distribución Dickey-Fuller. Estos valores críticos dependen del tamaño de la muestra y del nivel de significancia elegido.

7. **Conclusión:**

* Si el valor del estadístico de prueba es menor que el valor crítico correspondiente, se rechaza la hipótesis nula $H_0$, lo que implica que la serie temporal es estacionaria.

* Si el valor del estadístico de prueba es mayor que el valor crítico correspondiente, no se puede rechazar la hipótesis nula $H_0$, lo que implica que la serie temporal tiene raíces unitarias y es no estacionaria.

#**Dickey-Fuller Aumentada (ADF)**

La prueba de Dickey-Fuller Aumentada (ADF) es una extensión de la prueba de Dickey-Fuller que incluye términos adicionales para tener en cuenta la posible presencia de tendencias y autocorrelaciones en la serie temporal. La ecuación del modelo ADF es la siguiente:


$$Δ Y(t) = \alpha + \beta(t) + \gamma \cdot Y(t-1) + \delta_1 \cdot \Delta Y(t-1) + \ldots + \delta_p \cdot \Delta Y(t-p) + \varepsilon(t)$$

Donde:
$\Delta Y(t)$ es la primera diferencia de la serie temporal (es decir, la diferencia entre dos observaciones consecutivas)

$\alpha$ y $\beta$ son los términos de tendencia lineal

$\gamma\$ es el coeficiente de autoregresión

$\delta_1$, $\ldots$, $\delta_p$ son los coeficientes de las diferencias rezagadas de la serie temporal

$\varepsilon(t)$ es el término de error en el tiempo $t$.

La hipótesis nula de la prueba de Dickey-Fuller Aumentada es similar a la de la prueba de Dickey-Fuller: la presencia de una raíz unitaria en la serie temporal. El estadístico de prueba ADF se calcula de manera similar al de la prueba de Dickey-Fuller, pero teniendo en cuenta los términos adicionales en el modelo ADF.

En ambas pruebas, se utiliza el valor p para evaluar la significancia estadística de los resultados. Si el valor p es menor que un umbral predefinido (por ejemplo, 0.05), se rechaza la hipótesis nula y se concluye que la serie temporal es estacionaria.


##**Implementación de las pruebas DF y ADF**

  Se ejemplificará el procedimiento antes mencionado con los datos históricos del **los precios de cierre diarios del tipo de cambio del peso mexicano frente al dólar** utilizando la biblioteca estadística "statsmodels" en Python.

>   Importar las librerías necesarias:






```python
import pandas as pd #manipular y analizar datos tabulares
import numpy as np #manipulación de matrices y arreglos multidimensionales
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller # Esta función se utiliza para realizar la prueba de Dickey-Fuller para la estacionariedad en una serie de tiempo dada.
import yfinance as yf # descargar datos históricos y en tiempo real de instrumentos financieros desde Yahoo Finance
import matplotlib.pyplot as plt
```

>   Cargar los datos en forma de una serie de tiempo de los precios de cierre diarios del tipo de cambio del peso mexicano frente al dólar.

```python
# Descargar datos de Yahoo Finance
data = yf.download('MXN=X', start='2003-12-01', end='2023-07-13')

# Obtener los precios de cierre ajustados
USD_MXN = data['Adj Close']

#Mostrar datos cargados
print(USD_MXN)
```

```python
#Determine rolling statistics
media_movil = USD_MXN.rolling(window=90).mean() #window size 12 denotes 12 DAYS, giving rolling mean at yearly level
desviación_estándar = USD_MXN.rolling(window=90).std()
#print(rolmean,rolstd)

#Plot rolling statistics
plt.figure(figsize=(15, 15))
orig = plt.plot(USD_MXN, color='blue', label='USD/MXN')
mean = plt.plot(media_movil, color='red', label='Media móvil')
std = plt.plot(desviación_estándar, color='black', label='Desviación estándar')
plt.legend(loc='best')
plt.title('Serie de Tiempo de Precios de Cierre Ajustados y médias móviles - MXN=X ')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre Ajustado')
plt.show(block=False)
```

La idea principal de funcionamiento de las medias móviles es reducir el ruido o las fluctuaciones aleatorias presentes en una serie de tiempo, lo que permite identificar patrones o tendencias más suavizadas. Al calcular la media de un conjunto de observaciones consecutivas, se crea un nuevo punto de datos que representa un promedio de esos valores y que se coloca en el punto medio del intervalo considerado.

Se usará la función *adfuller* de la biblioteca *statsmodels* para realizar el test de Dickey-Fuller (DF).

```python
def Dickey_Fuller(serie_tiempo):
    # Aplica el test de Dickey-Fuller
    resultado = adfuller(serie_tiempo)

    # Extrae los valores del resultado
    estadistico_DF = resultado[0] #extrae el valor del estadístico de prueba DF
    p_value = resultado[1] #extrae el valor p del resultado
    valores_criticos = resultado[4] #extrae los valores críticos del resultado

    # Imprime los resultados
    print("Resultados de la prueba Dickey-Fuller:")
    print(f"estadistico_Dickey-Fuller: {estadistico_DF}") # imprime el valor del estadístico de prueba
    print(f"p-value: {p_value}")
    print("Valores críticos:")

    #itera sobre los valores críticos y los imprime
    for key, value in valores_criticos.items():
        print(f"\t{key}: {value}")

    # Realiza la interpretación estadística
    if p_value < 0.05:
        print("La serie es estacionaria (rechazo de la hipótesis nula)")
    else:
        print("La serie no es estacionaria (rechazo de la hipotesis alternativa)")
```

Mandar los datos de la serie de tiempo a la función y mostrar los resultados.

```python
Dickey_Fuller(USD_MXN);
```

El valor p (p-value)  0.6075599905071544, representa la probabilidad de obtener el estadístico de prueba observado.

Valores críticos (Critical Values), comparando tres niveles de confianza con el valor del estadístico de prueba. Dado que el valor del estadístico de prueba es más negativo que estos valores críticos, se puede rechazar la hipótesis nula de no estacionariedad.

Se usará la función *adfuller* de la biblioteca *statsmodels* para realizar el test de Dickey-Fuller Aumentada (ADF).

```python
def ADF(data):
    result = adfuller(data, autolag='AIC')

    print('Prueba de Dickey-Fuller Aumentada:')
    print('Estadístico de prueba:', result[0])
    print('Valor p:', result[1])
    print('Valores críticos:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
```

Para el pronóstico de la serie de tiempo no estacionaria se aplicará el método de promedios móviles exponenciales (EMA)

```python
#Importamos librería
from statsmodels.tsa.holtwinters import ExponentialSmoothing

#columna que represente el índice temporal
data.reset_index(inplace=True)
```

Se dividirán los datos en conjuntos de entrenamiento y prueba.

Utilizaremos el 80% de los datos para entrenar el modelo y el 20% restante para evaluar el rendimiento del pronóstico.

```python
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]
#train_data
```

 Ajustamos el modelo EMA a los datos de entrenamiento y realizamos el pronóstico.

```python
#model = ExponentialSmoothing(train_data['Close'], trend='add', seasonal='add')
model = ExponentialSmoothing(train_data['Close'], trend='add', seasonal='add', seasonal_periods=30)

fitted_model = model.fit()
forecast = fitted_model.forecast(len(test_data))
#print(test_data)
```

Evalúa el rendimiento del pronóstico.

```python
mse = ((forecast - test_data['Close']) ** 2).mean()
mae = (forecast - test_data['Close']).abs().mean()

print('Error cuadrado medio (MSE):', mse)
print('Error absoluto medio (MAE):', mae)
```

Se usa un modelo de pronóstico ARIMA(Autoregressive Integrated Moving Average), para hacer predicciones futuras.

```python
import statsmodels.api as sm

#Sea 100 el número de pasos de tiempo anteriores que se utilizan como predictores para el valor actual de la serie.
#5 Representa el número de veces que se realiza la diferenciación para convertir la serie en estacionaria.
#3 número de términos de promedio móvil que se utilizan para modelar la relación entre los errores del modelo y los valores pasados de la serie.

model = sm.tsa.ARIMA(test_data['Adj Close'], order=(60, 1, 2))
model_fit = model.fit()
```
